```python
import json
import matplotlib.pyplot as plt
import os
```

```python
log_path = "../src/logs/srcnn_history.json"

with open(log_path, "r") as f:
    history = json.load(f)

print(history.keys())
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Loss curve
axes[0].plot(history["train_loss"], label="Train Loss")
axes[0].plot(history["val_loss"], label="Val Loss")
axes[0].set_title("Loss")
axes[0].set_xlabel("Epoch")
axes[0].set_ylabel("MSE Loss")
axes[0].legend()
axes[0].grid(True)

# PSNR Curve
axes[1].plot(history["val_psnr"], label="Val PSNR", color="orange")
axes[1].set_title("PSNR over epochs")
axes[1].set_xlabel("Epoch")
axes[1].set_ylabel("PSNR (dB)")
axes[1].grid(True)

# SSIM Curve
axes[2].plot(history["val_ssim"], label="Val SSIM", color="green")
axes[2].set_title("SSIM over epochs")
axes[2].set_xlabel("Epoch")
axes[2].set_ylabel("SSIM")
axes[2].grid(True)

plt.tight_layout()
plt.show()
```

```python
print(f"Final Train Loss: {history['train_loss'][-1]:.6f}")
print(f"Final Val Loss:   {history['val_loss'][-1]:.6f}")
print(f"Final Val PSNR:   {history['val_psnr'][-1]:.2f} dB")
print(f"Final Val SSIM:   {history['val_ssim'][-1]:.4f}")
```
